```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0,10.0)

# Reading Data
data = pd.read_csv('headbrain.csv')
print(data.shape)
print(data.head())
```

```python
#assigning the values of x and y from the above file
x = data['Head Size(cm^3)'].values
y = data[ 'Brain Weight(grams)'].values

#finding the mean of x and y 
x_mean = np.mean(x)
y_mean = np.mean(y)

# Total number of values
n = len(x)

# Calculating b0 and b1
numer = 0
denom = 0
for i in range(n):
    numer += (x[i] - x_mean)*(y[i] - y_mean)
    denom += (x[i] - x_mean)**2
b1 = numer/denom
b0 = y_mean - b1*x_mean

# Printing coefficients
print(b0, b1)
```

```python
# Ploting values on regression line

min_x = np.min(x) - 100
max_x = np.max(x) + 100

# Calculating regression line predicted values X, Y
X = np.linspace(min_x, max_x, 1000)
Y = b0 + b1*X

#Plotting regression line
plt.plot(X, Y, color = 'b', label = 'regression line')
#plottig Scatter points
plt.scatter(x, y, color = 'r', label='Scatter Plot')

plt.xlabel = ("Head Size in cm^3")
plt.ylabel = ("Brain Weight(grams)")
plt.legend()
plt.show()
```

```python
# Calculating R^2
p_m = 0
a_m = 0
for i in range(n):
    pred_y = b0 + b1*x[i]
    p_m += (y[i] - pred_y)**2
    a_m += (y[i] - y_mean)**2
r = 1-(p_m/a_m)
print(r)
```

```python
# Finding r2 using Scikit learn 

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x = x.reshape((n, 1))

reg = LinearRegression()
reg = reg.fit(x, y)

y_pred = reg.predict(x)

r2 = reg.score(x, y)
print(r2)
```
